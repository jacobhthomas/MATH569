Now we 
1. Loop through each column in the DataFrame
2. Count the number of NaN values 
3. Forward fill and backward fill to handle NaN values
4. Check if there are still NaN values after ffill and bfill
5. Fill the remaining NaNs with the most common value



```python
for column in local_traffic_merged.columns:
    nan_count_before = local_traffic_merged[column].isnull().sum()
    if nan_count_before > 0:
        print(f"Processing column: {column}")

        local_traffic_merged[column] = local_traffic_merged[column].ffill().bfill()

        most_common_value = local_traffic_merged[column].mode()[0]

        nan_count_after = local_traffic_merged[column].isnull().sum()
        if nan_count_after > 0:
            local_traffic_merged[column].fillna(most_common_value, inplace=True)

        if local_traffic_merged[column].isnull().any():
            raise Exception(f"NaN values still present in {column} after filling!")

        unique_values = local_traffic_merged[column].unique()[:10]
        nan_filled = nan_count_before - nan_count_after

        print(f"Used '{most_common_value}' to fill {nan_filled} NaNs in {column}")
        print(f"First 10 unique values in {column} after filling NaNs: {unique_values}\n")
```

```python
for column in local_traffic_merged.columns:
    if local_traffic_merged[column].isnull().any():
        # If NaN values are found, raise an exception
        raise Exception(f"NaN values found in column {column}")
```

## PCA Part

#### First step: standardizing the data using StandardScalar

`StandardScaler` in `scikit-learn` standardizes features by removing the mean and scaling to unit variance. This process is often called z-score normalization. Here's how it works step by step:

1. **Calculate the Mean and Standard Deviation**:
   For each feature in the dataset, `StandardScaler` calculates the mean (``) and standard deviation (``).

2. **Subtract the Mean**:
   From each feature value (`x`), subtract the mean of that feature. This centers the distribution of the feature around zero.

   \( z = x -  \)

3. **Divide by the Standard Deviation**:
   Divide the result by the standard deviation of the feature to scale the distribution to unit variance.

   \( z = \frac{x - }{} \)

After this process, each feature will have a mean of 0 and a standard deviation of 1, assuming a normally distributed feature.

##### Why Scaling?

**PCA Specifics**:
   In PCA, features are considered as directions in space. If one feature has a much larger scale than others, PCA could end up being biased towards those features. Standardizing ensures that the principal components are influenced equally by all features based on their variance, not their absolute scale.


```python
from sklearn.preprocessing import StandardScaler

# Select features to include in PCA and the predictive model (exclude non-feature columns as necessary)
features = local_traffic_merged.drop(['SPEED', 'TIME', 'accurate_time', 'BUS COUNT', 'MESSAGE COUNT',], axis=1)

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

### Apply PCA:

```python
from sklearn.decomposition import PCA

# Initialize PCA, you can choose the number of components e.g., 0.95 for the amount of variance you want to retain
pca = PCA(n_components=0.95)

# Fit PCA on the standardized data and transform the data
principal_components = pca.fit_transform(scaled_features)

# See how many components were selected to retain the variance
print(f"Number of components chosen: {pca.n_components_}")
```

```python
np.random.seed(569)

test_proportion = 0.2
# Test/train splits, with the count at the end to make sure we have the same number of indices at the end
permutation = np.random.permutation(range(len(local_traffic_merged)))
test_indices = permutation[:int(test_proportion * len(local_traffic_merged))]
train_indices = permutation[int(test_proportion * len(local_traffic_merged)):]
np.unique(len(test_indices)+len(train_indices))

# Our test dataframe is test_df, and the train dataframe is train_df
test_df = local_traffic_merged.iloc[test_indices]
train_df = local_traffic_merged.iloc[train_indices]
```

```python
# Use the indices to split the principal_components array and SPEED column for training and testing
X_train_pca = principal_components[train_indices]
X_test_pca = principal_components[test_indices]

y_train = local_traffic_merged.loc[train_indices, 'SPEED']
y_test = local_traffic_merged.loc[test_indices, 'SPEED']
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_pca, y_train)

r_squared = model.score(X_test_pca, y_test)
print(f"Model R^2 Score: {r_squared}")

# Calculate MSE and MAE
predictions = model.predict(X_test_pca)
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```
