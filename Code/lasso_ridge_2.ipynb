```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import scipy.stats as stats
import statsmodels.api as sm
```

```python
train = pd.read_csv('csvs/train_df.csv')
test = pd.read_csv('csvs/test_df.csv')
train
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_train_lasso = train.drop(['SPEED'], axis=1)
y_train = train['SPEED']
X_test_lasso = test.drop(['SPEED'], axis=1)
y_test = test['SPEED']
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_lasso, y_train)
lasso_features = X_train_lasso.loc[:, lasso.coef_ != 0]
```

```python
model_lasso = LinearRegression()
model_lasso.fit(lasso_features, y_train)

print("For MLR + LASSO:")
r_sqr = model_lasso.score(X_test_lasso, y_test)
print(f'R^2: {r_sqr}')

pred = model_lasso.predict(X_test_lasso)
mse = mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```
