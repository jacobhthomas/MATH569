**Applying PCA**

```python
import pandas as pd

train_df = pd.read_csv('csvs/train_df.csv')
train_df.head()
test_df = pd.read_csv('csvs/test_df.csv')
test_df.head()
```

Removing the categorical variables

```python
# Columns to be dropped
columns_to_drop = ['is_snowovercast', 'is_snowrainovercast', 'is_clear', 'is_overcast', 'is_partcloud', 'is_rainovercase']

train_df.drop(columns=columns_to_drop, inplace=True)
test_df.drop(columns=columns_to_drop, inplace=True)
# Display the modified DataFrame to confirm the changes
train_df.head()
test_df.head()
```

PCA assumes that the features are linearly independent. The target variable, often dependent on the features, should not be included as it could bias the PCA results.

```python
# Drop 'SPEED' from features and store it separately for both datasets
features_train = train_df.drop(columns=['SPEED'])
target_train = train_df['SPEED']
features_test = test_df.drop(columns=['SPEED'])
target_test = test_df['SPEED']
```

```python
from sklearn.decomposition import PCA
import numpy as np

features_train_components = features_train.copy()

pca = PCA()
pca.fit(features_train_components)  # Only fit PCA on the training data features

cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
print("Components to keep:")
print(d)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(cumsum, linewidth=3)
plt.axvline(x=d, color='r', linestyle='--', label=f'95% variance at {d} components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Components')
plt.legend()
plt.grid(True)
plt.show()
```

- **PCA Dimensionality Reduction**: 
  - `fit_transform` on training data:
    - **Purpose**: Learns the PCA transformation (i.e., determining the principal components) and applies it to reduce the dimensionality.
    - **Result**: Retains only the top 'd' principal components that explain 95% of the variance.
  - `transform` on testing data:
    - **Purpose**: Applies the same PCA transformation learned from the training data.
    - **Why**: Ensures that both training and testing data are transformed consistently, avoiding data leakage.
  Source: [StandardScaler.fit_transform documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html?source=post_page-----78f915cf96fe--------------------------------#sklearn.preprocessing.StandardScaler.fit_transform)
- **Conversion to DataFrame**:
  - Transformed data is converted back into pandas DataFrames with columns named as `PC1`, `PC2`, ..., up to `PCd`.

```python
train_features_reduced = pca.fit_transform(features_train)[:,:d]  # Apply PCA reduction
test_features_reduced = pca.transform(features_test)[:,:d]

# Convert the PCA-transformed data back into DataFrame
train_df_reduced = pd.DataFrame(train_features_reduced, columns=[f'PC{i+1}' for i in range(d)])
test_df_reduced = pd.DataFrame(test_features_reduced, columns=[f'PC{i+1}' for i in range(d)])

# Re-add 'SPEED' to both the reduced training and testing DataFrames
train_df_reduced['SPEED'] = target_train
test_df_reduced['SPEED'] = target_test
```

Linear Regression!

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model using the training data
model.fit(train_df_reduced.drop('SPEED', axis=1), train_df_reduced['SPEED'])

# Predict using the testing data
y_pred = model.predict(test_df_reduced.drop('SPEED', axis=1))

mse = mean_squared_error(test_df_reduced['SPEED'], y_pred)
r2 = r2_score(test_df_reduced['SPEED'], y_pred)
print("Mean Squared Error:", mse)
print("R-squared Value:", r2)
```

```python
import matplotlib.pyplot as plt

# Scatter plot for actual vs. predicted speeds
plt.figure(figsize=(10, 6))
plt.scatter(test_df_reduced['SPEED'], y_pred, alpha=0.7, color='blue', label='Predicted vs. Actual')
plt.xlabel('Actual Speed')
plt.ylabel('Predicted Speed')
plt.title('Comparison of Actual and Predicted Speeds')
plt.plot([test_df_reduced['SPEED'].min(), test_df_reduced['SPEED'].max()],
         [test_df_reduced['SPEED'].min(), test_df_reduced['SPEED'].max()], 'k--', lw=2, label='Perfect Fit')
plt.legend()
plt.grid(True)
plt.show()
```
